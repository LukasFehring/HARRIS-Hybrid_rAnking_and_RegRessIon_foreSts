```python
scenario = ASlibScenario()
fold = 1
amount_of_training_instances = 1
test_scenario_name = "CPMP-2015"
scenario.read_scenario(os.path.join("data", "aslib_data-master", test_scenario_name))

scenario.performance_data = scenario.performance_data.iloc[:len(scenario.performance_data) // 10*9]
scenario.feature_data = scenario.feature_data.iloc[:len(scenario.feature_data) // 10*9]


basinc_ranking_loss = corrected_spearman_footrule
basic_regression_loss = regression_error_loss
basic_borda_score = borda_score_mean_ranking
basic_impact_factor = 0.5
basic_stopping_criterion = max_depth
binary_decision_tree: BinaryDecisionTree = BinaryDecisionTree(basinc_ranking_loss, basic_regression_loss, basic_borda_score, basic_impact_factor, basic_stopping_criterion, stopping_threshold=3)
binary_decision_tree.loss_overview=[]

```
